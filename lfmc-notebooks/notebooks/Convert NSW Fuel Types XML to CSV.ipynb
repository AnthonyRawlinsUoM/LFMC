```python
import xml.etree.ElementTree as ET
import csv

tree = ET.parse("/media/arawlins/Backups/DataSources/geoserver_data/FuelModels/Fuel_All_States/NSW/nsw_FuelType_20151013.xml")
root = tree.getroot()

# open a file for writing

_data = open('/tmp/fuel_types_LUT.csv', 'w')

# create the csv writer object

csvwriter = csv.writer(_data)
_head = []

count = 0
for member in root.findall('Fuel_Type'):
    fuel_type = []
    
    if count == 0:
        Id = member.find('Id').tag
        _head.append(Id)

        Type_Name = member.find('Type_Name').tag
        _head.append(Type_Name)

        Description = member.find('Description').tag
        _head.append(Description)

        Surface_r = member.find('Surface_r').tag
        _head.append(Surface_r)

        Surface_k = member.find('Surface_k').tag
        _head.append(Surface_k)

        Surface_c = member.find('Surface_c').tag
        _head.append(Surface_c)

        Elevated_r = member.find('Elevated_r').tag
        _head.append(Elevated_r)

        Elevated_k = member.find('Elevated_k').tag
        _head.append(Elevated_k)

        Elevated_c = member.find('Elevated_c').tag
        _head.append(Elevated_c)

        Bark_r = member.find('Bark_r').tag
        _head.append(Bark_r)

        Bark_k = member.find('Bark_k').tag
        _head.append(Bark_k)

        Bark_c = member.find('Bark_c').tag
        _head.append(Bark_c)

        Wind_Reduction_Factor = member.find('Wind_Reduction_Factor').tag
        _head.append(Wind_Reduction_Factor)

        csvwriter.writerow(_head)
        count = count + 1

    Id = member.find('Id').text
    fuel_type.append(Id)

    Type_Name = member.find('Type_Name').text
    fuel_type.append(Type_Name)

    Description = member.find('Description').text
    fuel_type.append(Description)

    Surface_r = member.find('Surface_r').text
    fuel_type.append(Surface_r)

    Surface_k = member.find('Surface_k').text
    fuel_type.append(Surface_k)

    Surface_c = member.find('Surface_c').text
    fuel_type.append(Surface_c)

    Elevated_r = member.find('Elevated_r').text
    fuel_type.append(Elevated_r)

    Elevated_k = member.find('Elevated_k').text
    fuel_type.append(Elevated_k)

    Elevated_c = member.find('Elevated_c').text
    fuel_type.append(Elevated_c)

    Bark_r = member.find('Bark_r').text
    fuel_type.append(Bark_r)

    Bark_k = member.find('Bark_k').text
    fuel_type.append(Bark_k)

    Bark_c = member.find('Bark_c').text
    fuel_type.append(Bark_c)

    Wind_Reduction_Factor = member.find('Wind_Reduction_Factor').text
    fuel_type.append(Wind_Reduction_Factor)

    csvwriter.writerow(fuel_type)
_data.close()
```
