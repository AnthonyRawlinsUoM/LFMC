```python
import xarray as xr
import numpy as np

import matplotlib.pyplot as plt

nx = 2
nt = 4
field3d = np.random.rand(nt, nx, nx)
field2d = np.random.rand(nx, nx)
```

```python
field2d
```

```python
field3d
```

```python
plt.imshow(field2d, cmap='gray')
plt.show()
```

```python
greyscale_mask = np.array(np.asarray(field2d) * 255 + 0.5, np.uint8)
plt.imshow(greyscale_mask, cmap='gray')
plt.show()
greyscale_mask
```

```python
da = xr.DataArray(field3d)
```

```python
for d in da['dim_0']:
    plt.imshow(da.sel(dim_0=d))
    plt.show()
```

```python
da = da.where(greyscale_mask != 0) * (greyscale_mask/255)
```

```python
for d in da['dim_0']:
    plt.imshow(da.sel(dim_0=d))
    plt.show()
```
