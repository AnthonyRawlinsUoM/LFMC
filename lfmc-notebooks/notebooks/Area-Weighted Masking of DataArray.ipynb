```python
## Dead Fuel Model applicable only! - Affine suits VP3pm inputs only atm!
affine = Affine(0.05, 0.0, 111.975, 0.0, -0.05, -9.974999999999994)

def get_shapes_from_shp_file(shp):
    selections = list()
    shape = fiona.open(shp)
    for f in shape:
        selections.append(s)
    return selections

def get_shapes_from_geojson(geojson):
    selections = list()
    for f in geojson["features"]:
        if f["geometry"]["type"] is "Polygon":
            points = f["geometry"]["coordinates"]
            s = shapely.geometry.Polygon(*points)
            selections.append(s)
    return selections

def get_regionmask_from_geojson(geojson):
    selections = list()
    count = 0
    numbers = []
    names = []
    abbrevs = []
    if geojson["type"] == "FeatureCollection":
        for p in geojson["features"]:
            if p["geometry"]["type"] is "Polygon":
                points = p["geometry"]["coordinates"]
                s = shapely.geometry.Polygon(*points)
                selections.append(s)

                numbers.append(count)
                if p["properties"] == {}:
                    names.append("Selection_%s" % count)
                    abbrevs.append("SEL_%s" % count)

                count += 1
    else:
        if geojson["geometry"]["type"] is "Polygon":
                points = geojson["geometry"]["coordinates"]
                s = shapely.geometry.Polygon(*points)
                selections.append(s)

                numbers.append(count)
                if geojson["properties"] == {}:
                    names.append("Selection_%s" % count)
                    abbrevs.append("SEL_%s" % count)
                else:
                    names.append(geojson["properties"]["FIRENAME"])
                    abbrevs.append("SEL_%s" % count)
                count += 1
    
    return regionmask.Regions_cls('Selections', numbers, names, abbrevs, selections)

def get_bbox(poly):
    return list((poly.bounds[0], poly.bounds[2], poly.bounds[1], poly.bounds[3]))

def get_weighted_mask(regionmask):
    return None

def hull_shape(hull):
    return shapely.geometry.Polygon([hull.points[vertex] for vertex in hull.vertices])

def get_buffered_query_extents(regionmask, kilometers, plot=False):
    hull = get_query_hull(regionmask)
    
    if plot:
        plt.plot(points[:,0], points[:,1], 'o')
        for simplex in hull.simplices:
            plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
        plt.show()
    
    buffered_hull_shp = as_buffered(hull_shape(hull), 10)
    hull_bbox = get_bbox(buffered_hull_shp)
    
    return hull_bbox

def get_query_hull(regionmask):
    points = np.concatenate(regionmask.coords, axis=0)
    return ConvexHull(points)
    
def get_corners(poly):
    bb = bounds_to_bbox(poly)
    return [(bb[0], bb[2]), (bb[0], bb[3]), (bb[1], bb[3]),(bb[1], bb[2])]

def get_buffered_coords(poly, kilometers):
    return transform_to_meters(poly).buffer(kilometers)
    
def transform_to_meters(poly):
    new_points = [~affine * (point) for point in asarray(poly.exterior)]
#     print(new_points)
    return shapely.geometry.Polygon(new_points)

def transform_to_latlong(poly):
    return shapely.geometry.Polygon([affine * (point) for point in asarray(poly.exterior)])

def as_buffered(poly, kilometers):
    return transform_to_latlong(get_buffered_coords(poly, kilometers))

def weighted_mask(query_poly):
    c = Line.Contour(query_poly)
    raster = Rasterizer(c, 886, 691).get()
    raster = np.array(np.asarray(raster)*255+0.5, np.uint8)
    return raster

def fast_weighted_mask(query_poly):
    c = Line.Contour(query_poly)
    raster = Rasterizer(c, 886, 691).get_fast()
    raster = np.array(np.asarray(raster)*255+0.5, np.uint8)
    return raster

def transform_selection(regionmask_obj):
    aff = Affine(0.05, 0.0, 111.975, 0.0, -0.05, -9.974999999999994)
    points = np.concatenate(regionmask_obj.coords, axis=0)
    all_points = [(u,v) for [u,v] in points]
    _pixel_coords = [~aff * (coord) for coord in all_points]
    return _pixel_coords
```
