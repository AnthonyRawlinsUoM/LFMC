# Convenience Script to generate N colors from Viridis Colormap

```python
from pylab import *

def createSLD(name, rs, rf, is_index, reverse):
    
    
    if is_index:
        r = (rf - rs)
        levels = r + 1
    else:
        r = (rf - rs)
        levels = 255

    cmap = cm.get_cmap('viridis', levels)

    if reverse:
        rev = 'Reversed'
    else:
        rev = ''

    stepsize = r/(levels-1)
    result = ''
    
    result += '<?xml version="1.0" encoding="ISO-8859-1"?>\n\
    <StyledLayerDescriptor version="1.0.0"\n\
        xsi:schemaLocation="http://www.opengis.net/sld StyledLayerDescriptor.xsd"\n\
        xmlns="http://www.opengis.net/sld"\n\
        xmlns:ogc="http://www.opengis.net/ogc"\n\
        xmlns:xlink="http://www.w3.org/1999/xlink"\n\
        xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance">\n\
      <NamedLayer>\n\
        <Name>'+name +'_Viridis_' + str(levels) +'lvls_'+ str(r) +'rng_'+ rev + '</Name>\n\
        <UserStyle>\n\
          <Title>SLD Viridis</Title>\n\
          <FeatureTypeStyle>\n\
            <Rule>\n\
              <RasterSymbolizer>\n\
                <ColorMap>\n'

    for i in range(cmap.N):
        if reverse:
            rgb = cmap(levels-(i+1))[:3] # will return rgba, we take only first 3 so we get rgb
        else:
            rgb = cmap(i)[:3] # will return rgba, we take only first 3 so we get rgb
        result += "                    <ColorMapEntry color=\"%s\" quantity=\"%1.7f\" />\n" % (matplotlib.colors.rgb2hex(rgb), (i*stepsize)+rs)
    result += '                  </ColorMap>\n\
              </RasterSymbolizer>\n\
            </Rule>\n\
          </FeatureTypeStyle>\n\
        </UserStyle>\n\
      </NamedLayer>\n\
    </StyledLayerDescriptor>'
        
    return result
```

```python
# Indexes
dfct = createSLD(name='DF', rs=0, rf=10, is_index=True, reverse=False)         # Wet -> Dry
ffdi = createSLD(name='FFDI', rs=0, rf=8, is_index=True, reverse=False)        # Wet -> Dry
gfdi = createSLD(name='GFDI', rs=0, rf=8, is_index=True, reverse=False)        # Wet -> Dry
kbdi = createSLD(name='KBDI', rs=0, rf=200, is_index=True, reverse=False)      # Wet -> Dry

# Continuous
jasm = createSLD(name='JASMIN', rs=0, rf=204, is_index=False, reverse=False)   # Wet -> Dry
curi = createSLD(name='Curing', rs=0, rf=100, is_index=False, reverse=False)   # Wet -> Dry
dfmc = createSLD(name='DFMC', rs=6, rf=35, is_index=False, reverse=True)     # Dry -> Wet
awra = createSLD(name='AWRA', rs=0, rf=1, is_index=False, reverse=True)        # Wet -> Dry
lfmc = createSLD(name='LFMC', rs=0, rf=200, is_index=False, reverse=False)     # Dry -> Wet

lvmc = createSLD(name='LVMC', rs=2, rf=305, is_index=False, reverse=True)     # Dry -> Wet
# matt = createSLD(name='Mattthews', rs=0, rf=200, is_index=False, reverse=False)     # Wet-> Dry
# temp = createSLD(name='Temp', rs=0, rf=45, is_index=False, reverse=False)   # Wet -> Dry
```
